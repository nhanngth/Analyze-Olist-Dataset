```python
state = df1['State']
city = df1['City']
municipality = df2['Municipality']
large_city = df3['Largest city']
```

```python
all_bra = pd.concat([state,city,municipality,large_city],ignore_index=True)
```

```python
all_bra = all_bra.str.lower()
```

```python
all_bra = all_bra.drop_duplicates()
```

```python
all_bra = all_bra.sort_values().reset_index(drop=True)
```

```python
check = pd.DataFrame(all_bra,columns=['city'])
```
