Cleaning Reviews table

```python
reviews = pd.read_csv("/content/olist_order_reviews_dataset.csv")
```

```python
reviews.head()
```

```python
reviews.info()
```

```python
reviews['review_creation_date'] = pd.to_datetime(reviews['review_creation_date'], format='%Y-%m-%d')
```

```python
reviews.info()
```

```python
reviews['review_comment_title'].sort_values(ascending=False).head(20)
```

```python
def emoji_replace(text):
    for emoji in UNICODE_EMOJI:
        text = text.replace(emoji, str(' ') + '_'.join(UNICODE_EMOJI[emoji].replace(',','').replace(':','').split())).strip()
    return text
```

```python
for i in range(0,len(reviews['review_comment_title'])):
    if reviews.loc[i,'review_comment_title'] != np.nan:
        reviews.loc[i,'review_text'] = emoji_replace(str(reviews.loc[i,'review_comment_title']))
    else: reviews.loc[i,'review_text'] = np.nan
```

```python
for i in range(0,len(reviews['review_comment_message'])):
    if reviews.loc[i,'review_comment_message'] != np.nan:
        reviews.loc[i,'review_comment_text'] = emoji_replace(str(reviews.loc[i,'review_comment_message']))
    else: reviews.loc[i,'review_comment_text'] = np.nan
```

```python
del reviews['review_comment_title']
del reviews['review_comment_message']
```

```python
reviews
```

```python
reviews = reviews[['review_id', 'order_id', 'review_score', 'review_text', 
         'review_comment_text','review_creation_date', 'review_answer_timestamp']]
```

```python
reviews.loc[reviews['review_text'] == 'nan','review_text'] = ''
reviews.loc[reviews['review_comment_text'] == 'nan','review_comment_text'] = ''
```

```python
reviews.to_csv(r'/content/olist_output/reviews.csv', index=False)
```
